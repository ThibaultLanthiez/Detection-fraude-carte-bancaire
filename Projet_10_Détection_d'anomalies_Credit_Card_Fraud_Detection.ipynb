```
# Import de bibliothèques

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```
# Import de données (via Kaggle : https://www.kaggle.com/mlg-ulb/creditcardfraud)
data = pd.read_csv('/content/creditcard.csv')

print(f'Le dataset contient {data.shape[0]} lignes et {data.shape[1]} colonnes.')
data
```

```
print(data['Class'].value_counts(), '\n')
print(data['Class'].value_counts(normalize=True)*100)

# 99.8 % des valeurs ne sont pas des fraudes
```

```
# Suppression de certaines colonnes
data.drop(data.tail(1).index, inplace=True)
data.drop(['Time'], axis=1, inplace=True)

data_class = data['Class']
data.drop(['Class'], axis=1, inplace=True)
```

```
# Création du jeu d'entraînement
y_train = data_class
X_train = data
```

```
# Import de bibliothèques de Machine Learning

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

import xgboost as xgb

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import learning_curve
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score
```

```
# Test de plusieurs algorithmes de Machine Learning (9 modèles)

# Ratio du nombre de fraudes
ratio = data_class[data_class == 1].shape[0] / (data.shape[0]-1)

print(f'{round(ratio,5)} de fraudes contre {data_class[data_class == 0].shape[0] / (data.shape[0]-1)} de non-fraudes.')
print(f'L\'objectif est donc de dépasser {data_class[data_class == 0].shape[0] / (data.shape[0]-1)}\n')

model_1 = AdaBoostClassifier(random_state=0)
model_2 = DecisionTreeClassifier(random_state=0)
model_3 = RandomForestClassifier(random_state=0)
model_4 = SVC(random_state=0)
model_5 = KNeighborsClassifier()
model_6 = LogisticRegression(random_state=0)
model_7 = GradientBoostingClassifier(random_state=0)
model_8 = BaggingClassifier(base_estimator=DecisionTreeClassifier(random_state=0),n_estimators=100)
model_9 = xgb.XGBClassifier(random_state=0)

dict_models = {'AdaBoostClassifier':model_1,
               'DecisionTreeClassifier':model_2,
               'RandomForestClassifier':model_3,
               'SVC':model_4,
               'KNeighborsClassifier':model_5,
               'LogisticRegression':model_6,
               'GradientBoostingClassifier':model_7,
               'BaggingClassifier':model_8,
               'XGBClassifier':model_9}

for key, model in dict_models.items():
  score = cross_val_score(model, X_train, y_train, cv=2).mean()*100
  print(f'- {key} : {score} %')

# Performances sur la validation croisée :
  # - AdaBoostClassifier : 99.907 %
  # - DecisionTreeClassifier : 99.827 %
  # - RandomForestClassifier : 99.927 %
  # - SVC : 99.867 %
  # - KNeighborsClassifier : 99.892 %
  # - LogisticRegression : 99.910 %
  # - GradientBoostingClassifier : 99.856 %
  # - BaggingClassifier : 99.914 %
  # - XGBClassifier : 99.921 %
```

```
from sklearn.ensemble import IsolationForest
# On arrête avec les algo de classificatio binaire et on passe aux algo de détection d'anomalies

ratio = data_class[data_class == 1].shape[0] / (data.shape[0]-1)
print(f'{round(ratio,5)} de fraudes contre {data_class[data_class == 0].shape[0] / (data.shape[0]-1)} de non-fraudes.')
print(f'L\'objectif est donc de dépasser {data_class[data_class == 0].shape[0] / (data.shape[0]-1)}')

model = IsolationForest(random_state=0, contamination=round(ratio,5))

model.fit(data)
pred = model.predict(data)

# outliers = pred == -1 
# a_garder = pred == 1

result = np.where(pred==1, 0, pred) 
result = np.where(result==-1, 1, result) 

res = (result == np.array(data_class)).mean()*100
print(f'- IsolationForest : {res} %') # Pas meilleur que les algo de classification
```

```
from sklearn.covariance import EllipticEnvelope
# Autres algo de détection d'anomalies

ratio = data_class[data_class == 1].shape[0] / (data.shape[0]-1)
print(f'{round(ratio,5)} de fraudes contre {data_class[data_class == 0].shape[0] / (data.shape[0]-1)} de non-fraudes.')
print(f'L\'objectif est donc de dépasser {data_class[data_class == 0].shape[0] / (data.shape[0]-1)}')

model = EllipticEnvelope(random_state=0, contamination=round(ratio,5))

model.fit(data)
pred = model.predict(data)

# outliers = pred == -1 
# a_garder = pred == 1

result = np.where(pred==1, 0, pred) 
result = np.where(result==-1, 1, result) 

res = (result == np.array(data_class)).mean()*100
print(f'- EllipticEnvelope : {res} %') # Moins bon que IsolationForest
```
