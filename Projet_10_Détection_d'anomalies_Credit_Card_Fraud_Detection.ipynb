```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('creditcard.csv')

print(f'Le dataset contient {data.shape[0]} lignes et {data.shape[1]} colonnes.')
data
```

```python
print(data['Class'].value_counts(), '\n')
print(data['Class'].value_counts(normalize=True)*100)
```

```python
data.drop(data.tail(1).index, inplace=True)
data.drop(['Time'], axis=1, inplace=True)

data_class = data['Class']
data.drop(['Class'], axis=1, inplace=True)
```

```python
y_train = data_class
X_train = data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import learning_curve
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score
```

```python
ratio = data_class[data_class == 1].shape[0] / (data.shape[0]-1)
print(f'{round(ratio,5)} de fraudes contre {data_class[data_class == 0].shape[0] / (data.shape[0]-1)} de non-fraudes.')
print(f'L\'objectif est donc de dépasser {data_class[data_class == 0].shape[0] / (data.shape[0]-1)}\n')

model_1 = AdaBoostClassifier(random_state=0)
model_2 = DecisionTreeClassifier(random_state=0)
model_3 = RandomForestClassifier(random_state=0)
#model_4 = SVC(random_state=0)
model_5 = KNeighborsClassifier()
model_6 = LogisticRegression(random_state=0)
model_7 = GradientBoostingClassifier(random_state=0)
model_8 = BaggingClassifier(base_estimator=DecisionTreeClassifier(random_state=0),n_estimators=100)

dict_models = {'AdaBoostClassifier':model_1,
               'DecisionTreeClassifier':model_2,
               'RandomForestClassifier':model_3,
               #'SVC':model_4,
               'KNeighborsClassifier':model_5,
               'LogisticRegression':model_6,
               'GradientBoostingClassifier':model_7,
               'BaggingClassifier':model_8}

for key, model in dict_models.items():
  val_score = cross_val_score(model, X_train, y_train, cv=2).mean()*100
  print(f'- {key} : {val_score} %')

# - AdaBoostClassifier : 99.90695420742541 %
# - DecisionTreeClassifier : 99.82654859799302 %
# - RandomForestClassifier : 99.92661671453551 %
# - SVC : 99.8672780770068 %
# - KNeighborsClassifier : 99.89185621089443 %
# - LogisticRegression : 99.90906090461577 %
# - GradientBoostingClassifier : 99.85639347485657 %
# - BaggingClassifier : 99.9136254151949 %
# - XGBClassifier : 99.9206477391628 %
```

IsolationForest

```python
from sklearn.ensemble import IsolationForest

ratio = data_class[data_class == 1].shape[0] / (data.shape[0]-1)
print(f'{round(ratio,5)} de fraudes contre {data_class[data_class == 0].shape[0] / (data.shape[0]-1)} de non-fraudes.')
print(f'L\'objectif est donc de dépasser {data_class[data_class == 0].shape[0] / (data.shape[0]-1)}')

model = IsolationForest(random_state=0, contamination=round(ratio,5))
model.fit(data)
pred = model.predict(data)

# outliers = pred == -1 
# a_garder = pred == 1

result = np.where(pred==1, 0, pred) 
result = np.where(result==-1, 1, result) 

res = (result == np.array(data_class)).mean()*100
print(f'- IsolationForest : {res} %')
# - IsolationForest : 99.75316531252852 %
```

EllipticEnvelope

```python
from sklearn.covariance import EllipticEnvelope

ratio = data_class[data_class == 1].shape[0] / (data.shape[0]-1)
print(f'{round(ratio,5)} de fraudes contre {data_class[data_class == 0].shape[0] / (data.shape[0]-1)} de non-fraudes.')
print(f'L\'objectif est donc de dépasser {data_class[data_class == 0].shape[0] / (data.shape[0]-1)}')

model = EllipticEnvelope(random_state=0, contamination=round(ratio,5))
model.fit(data)
pred = model.predict(data)

# outliers = pred == -1 
# a_garder = pred == 1

result = np.where(pred==1, 0, pred) 
result = np.where(result==-1, 1, result) 

res = (result == np.array(data_class)).mean()*100
print(f'- EllipticEnvelope : {res} %')
# - EllipticEnvelope : 99.67732421367528 %
```

LocalOutlierFactor

```python
from sklearn.neighbors import LocalOutlierFactor

ratio = data_class[data_class == 1].shape[0] / (data.shape[0]-1)
print(f'{round(ratio,5)} de fraudes contre {data_class[data_class == 0].shape[0] / (data.shape[0]-1)} de non-fraudes.')
print(f'L\'objectif est donc de dépasser {data_class[data_class == 0].shape[0] / (data.shape[0]-1)}')

nb_sup = data_class[data_class == 1].shape[0]

model = LocalOutlierFactor(n_neighbors=nb_sup)
pred = model.fit_predict(data)

# outliers = pred == -1 
# a_garder = pred == 1

result = np.where(pred==1, 0, pred) 
result = np.where(result==-1, 1, result) 

res = (result == np.array(data_class)).mean()*100
print(f'- LocalOutlierFactor : {res} %')
# - LocalOutlierFactor : 90.11853682857804 %
```
